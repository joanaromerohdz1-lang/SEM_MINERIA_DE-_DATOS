**FUNCIONES AVANZADAS - PANDAS (PART 1)**

```python
import pandas as pd
df = pd.read_csv('empleados.csv')
df.abs

#mostrar la info tecnica del dataframe
df.info()
```

**1-**Obtén todas las filas del depto "Eng" que además tengan perf_score >= 0.85, pero muestra solo las columnas: emp_id, name, city, perf_score, salary_mxn.

```python
df.loc[
    (df['dept'] == 'Eng') & (df['perf_score'] >= 0.85),
    ['emp_id', 'name', 'city', 'perf_score', 'salary_mxn']
]
```

**2-**Del DataFrame completo, obtén las filas 3 a 8 (incluyendo 3 y excluyendo 8), y solo las columnas 0, 1, 5 y 6 (por posición).

```python
df.iloc[3:8, [0, 1, 5, 6]]
```

**3-Calcula:**Filtra empleados que estén en city CDMX o GDL, con warnings == 0 y absences <= 1. Muestra emp_id, name, city, warnings, absences.

```python
df.loc[
    (df['city'].isin(['CDMX', 'GDL'])) &
    (df['warnings'] == 0) &
    (df['absences'] <= 1),
    ['emp_id', 'name', 'city', 'warnings', 'absences']
]
```

**4-Calcula:**
salario promedio (salary_mxn)
perf promedio (perf_score)
total de warnings (warnings)

```python
df.agg({
    'salary_mxn': 'mean',
    'perf_score': 'mean',
    'warnings': 'sum'
})
```

**5-Por cada dept, calcula:**
número de empleados
salario promedio
perf promedio
Ordena el resultado por salario promedio descendente.

```python
df.groupby('dept') \
  .agg(
      num_empleados=('emp_id', 'count'),
      salario_promedio=('salary_mxn', 'mean'),
      perf_promedio=('perf_score', 'mean')
  ) \
  .sort_values(by='salario_promedio', ascending=False)
```
